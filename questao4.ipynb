```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x**4 - 4*x**2 + 4

def df(x):
  return 4*x**3 - 8*x
```

```python
x = np.linspace(-100,100,100)
fx = f(x)
```

```python
plt.plot(x,fx)
plt.grid(True)
```

```python
estimativa_inicial = -25
```

```python
iteracoes = 0

while(iteracoes <= 5):
  raiz = estimativa_inicial - f(x)/df(x)
  estimativa_inicial = raiz
  iteracoes +=1
```

```python
print(f'iteracoes : {iteracoes}\n raiz: {raiz}')
```
